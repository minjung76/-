```python

#각 컬럼 기본 정도 확인, 데이터의 수(결측값 제외), 타입 확인
df.info()
```

```python
#null 값이 몇개인지 확인, 결측치 확인과정
df.isnull().sum()
```

```python
#결측치 처리를 어떻게 할것인지 고민
#방법 1 : 0방으로 처리하는 방법
#방법 2 : 평균으로 처리하는 법 mean()
#방법 2 : 표본의 중앙값으로 처리 median()
#방법 3 : 최대값으로 처리하는 방법 max()
#방법 4 : 최소값으로 처리하는 방법 min()
#방법 5 : 최빈값으로 처리하는 방법 mode
print(df["Age"].mean().round())
print(df["Age"].median())
print(df["Age"].max())
print(df["Age"].min())
print(df["Age"].mode())
```

```python
#Age 필드의 결측값 즉 null 값을 0으로 대체
df["Age"].fillna(0)
```

```python
#Age 필드의 결측값 즉 null 값을 평균값으로 대체
df["Age"].fillna(df["Age"].mean())
```

```python
df.isnull().sum()
```

```python
#Age 필드의 결측값 즉 null 값을 평균값으로 대체하고 원본 dataframe에 반영
df["Age"].fillna(df["Age"].mean(),inplace=True)
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(0)
```

```python
df.isnull().sum()
```

```python
#필드 value 값 확인
df["Embarked"].value_counts()
```

```python
df["Embarked"].value_counts().sort_index()
```

```python
#embark_town 열이 null 일 경우 해당 열을 삭제하고 데이터에 반영
df.dropna(subset=["Embarked"], axis=0, inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
import numpy as np
#조건으로 결측치 처리하기
#df['변수명'] = np.where(조건문 , 조건에 충족할 시 np.nan,원래 변수의 값)
#그 후 dropna(subset = [])를 이용해 결측치를 없애주는 방식으로 해줘야합니다. 

df['Age'] = np.where(df['Age'] == 3, np.nan , df['Age'])
df.dropna(subset = ['Age'], inplace=True)
```

```python
df.info()
```

```python
#결측값이 있는 모든 행을 제거
#df.dropna(axis=0,inplace=True)
#df.isnull().sum()
df.info()
```

```python
#df['name'] = np.where(df['name'] == '말포이', np.nan ,df['name']) 결측치로 안바뀌고 문자nan반환
#밑에 코드처럼 해야
#df['name'] = np.where(df['Embarked'] == 'Q', 'C' ,df['Embarked'])
df['Embarked'] = df['Embarked'].replace('Q','C') 
```
